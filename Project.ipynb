```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.liliorigin.com/products/5-pack-changeable-earring"
```

```python
page=requests.get(url)
print(page)
```

```python
data=page.text
```

```python
soup=bs(data,"html.parser")
```

```python
print(soup.prettify())
```

```python
title=soup.title
print(title.text)
```

# To get customer's names

```python
names=soup.find_all("span",class_="jdgm-rev__author")
print(names)
```

```python
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].text)
print(cust_name,end=" ")
```

# To get customer's comments

```python
comments=soup.find_all("b",class_="jdgm-rev__title")
print(comments)
```

```python
cust_comments=[]
for i in range(0,len(comments)):
    cust_comments.append(comments[i].text)
print(cust_comments)
```

# To get price of the product(w/o sale)

```python
price=soup.find("s",class_="price-item price-item--regular")
print(price.text)
```

```python
print(price)
```

# To get price of product(w sale)

```python
sale_price=soup.find("span",class_="price-item price-item--sale price-item--last")
print(sale_price)
```

# To get customer's review

```python
reviews=soup.find_all("div",class_="jdgm-rev__body")
print(reviews)
```

```python
cust_rev=[]
for i in range(0,len(reviews)):
    cust_rev.append(reviews[i].text)
print(cust_rev,end=" ")
```

# Histogram of percentage for each stars recieved 

```python
perstar=soup.find_all("div",class_="jdgm-histogram__percentage")
print(perstar)
```

```python
per_star=[]
for i in range(0,len(perstar)):
    per_star.append(perstar[i].text)  
print(per_star,end=" ")
```
